```python
import matplotlib.pyplot as plt
import numpy as np

def generate_age_data(sample_size=1000):
    # Generate random age data for demonstration
    np.random.seed(42)
    ages = np.random.randint(18, 65, sample_size)
    return ages

def plot_histogram(data, variable_name):
    plt.hist(data, bins=20, color='skyblue', edgecolor='black')
    plt.title(f'Distribution of {variable_name}')
    plt.xlabel(variable_name)
    plt.ylabel('Frequency')
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.show()

def main():
    # Generate random age data
    ages = generate_age_data()

    # Plot histogram
    plot_histogram(ages, 'Age')

if __name__ == "__main__":
    main()
```
